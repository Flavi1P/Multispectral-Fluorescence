# Prediction of the four different cluster
<br/>
We will try in this notebook to predict the four clusters in the hplc dataset from the 3x1m signal.

```python
import pandas as pd

hplc_fluo = pd.read_csv("/home/flavien/Documents/these/mf/Boussole/Output/Data/Compiled/hplc_mf_clusterised.csv")
hplc_fluo.head()
#hplc_fluo = hplc_fluo[hplc_fluo["cluster"] != "group_1" ]
```

```python
target_name = "cluster"
target = hplc_fluo[target_name]
target.value_counts()
```

```python
data = hplc_fluo[["fluo_440", "fluo_470", "fluo_532", "bb700"]]
data.describe()
```

Divide the dataset into learning and training test

```python
from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(
    data, target, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

model = Pipeline(steps=[
    ("preprocessor", StandardScaler()),
    ("classifier", KNeighborsClassifier(n_neighbors=5)),
])
```

```python
from sklearn.model_selection import cross_validate

cv_results = cross_validate(model, data, target, cv = 5)
scores = cv_results["test_score"]
print(scores)
```
